```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the dataset
file_name = "5G-NIDD/Combined.csv"
dataset = pd.read_csv(file_name)
dataset
```

```python
# Compute max and min for each feature
max_values = dataset.max()
min_values = dataset.min()

# Combine max and min into a single DataFrame
summary = pd.DataFrame({
    'Feature': dataset.columns,
    'Max Value': max_values.values,
    'Min Value': min_values.values
})

# Save the summary to a text file
summary.to_csv('feature_min_max_summary.txt', index=False, sep='\t')
```

```python
# Step 1: Remove duplicates
dataset = dataset.drop_duplicates()
dataset
```

```python
from sklearn.model_selection import train_test_split


# Step 3: Split dataset into training and testing sets
X = dataset.drop(['Label', 'Attack Type', 'Attack Tool', 'ID'], axis=1, errors='ignore')  # Features
y = dataset['Label']  # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler

# Step 4: Scaling the Dataset

# Select numeric columns only
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns

# Scale only numeric columns
scaler = MinMaxScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])

X_train
```

```python
X_test[numeric_cols] = scaler.fit_transform(X_test[numeric_cols])

X_test
```

```python
# Step 5: Handle missing values (Raplce with Nan with 0s)
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)
```

```python
from imblearn.over_sampling import SMOTENC

categorical_cols = dataset.select_dtypes(include=['object']).columns
categorical_cols = [col for col in categorical_cols if col not in ['Label', 'Attack Type', 'Attack Tool']]

# Ensure all categorical columns are strings
for col in categorical_cols:
    X_train[col] = X_train[col].astype(str)


categorical_features = [X_train.columns.get_loc(col) for col in categorical_cols]

#Step 6: Apply SMOTE for binary-class
smote_nc = SMOTENC(categorical_features=categorical_features, random_state=42)
X_train_resampled, y_train_resampled = smote_nc.fit_resample(X_train, y_train)
```

```python
# Check class distribution
print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", pd.Series(y_train_resampled).value_counts())

# Combine X_train_resampled and y_train_resampled into a single DataFrame
resampled_data = pd.DataFrame(X_train_resampled, columns=X_train.columns)  # Use original feature names
resampled_data['Label'] = y_train_resampled  # Add the target variable as a new column

# Combine X_test and y_test into a single DataFrame
test_data = pd.DataFrame(X_test, columns=X_test.columns)  # Use original feature names
test_data['Label'] = y_test  # Add the target variable as a new column

# Save to a CSV file
resampled_data.to_csv('Unified_Dataset/resampled_binary_training_data.csv', index=False)
test_data.to_csv('Unified_Dataset/binary_testing_data.csv', index=False)

print("Resampled data saved to 'resampled_binary_training_data.csv'")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Step 6: Train the Random Forest model
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_resampled, y_train_resampled)

# Step 7: Make predictions
y_pred = rf.predict(X_test)

# Step 8: Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
